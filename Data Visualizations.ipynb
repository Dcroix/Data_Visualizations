# Column Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create sample DataFrame
df = pd.DataFrame({
    'Category': ['Laptops', 'Desktops', 'Tablets', 'Smart_Phones', 'Television'],
    'Value': [23, 45, 12, 67, 34]
})

# Identify max value
max_index = df['Value'].idxmax()
max_value = df.loc[max_index, 'Value']

# Colors: neutral for all, highlight for the max
colors = ['#7F8FA3'] * len(df)        # neutral bars
colors[max_index] = '#77DDF9'         # highlight color (emphasized)

plt.figure(figsize=(8,5)) # this is for plot size only

# Set figure background color (outside the axes)
plt.gcf().patch.set_facecolor('#f7f7f7')  # light gray (optional)

ax = plt.gca()

# Set plot background color (the main area) - tip: make sure the colors go well together
ax.set_facecolor('#EAFBFF')  # soft bluish-gray â€” change to any color you like!

bars = plt.bar(df['Category'], df['Value'], color=colors)

plt.title('Gadgets Sold for Third Quarter', fontsize=16, weight='bold')
plt.xlabel('Category')
plt.ylabel('Units Sold')

# Add label ONLY to the highest bar
bar = bars[max_index]
plt.text(
    bar.get_x() + bar.get_width() / 2,
    max_value + 1,      # small gap above the bar
    str(max_value),
    ha='center',
    va='bottom',
    fontsize=11,
    fontweight='bold'
)

# Adjust Y-axis so the label never touches the top
y_padding = max_value * 0.15   # 15% headroom (adjust as you like)
plt.ylim(0, max_value + y_padding)

plt.tight_layout()
plt.show()
```

# Bar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame with MORE than 5 categories
df = pd.DataFrame({
    'Category': ['Alpha', 'Bravo', 'Charlie', 'Delta', 'Echo', 'Foxtrot', 'Golf'],
    'Value':    [34,      78,      22,        91,      45,     62,       55]
})

# Identify max value
max_index = df['Value'].idxmax()
max_value = df.loc[max_index, 'Value']

# Colors for bars (highlight only max)
colors = ['#F8D6D0'] * len(df)    # neutral
colors[max_index] = '#D97199'     # highlight

plt.figure(figsize=(9, 6))

ax = plt.gca()

# Background colors
plt.gcf().patch.set_facecolor('#f7f7f7')  # Figure background
ax.set_facecolor('#EAFBFF')               # Plot background

# Horizontal bar chart
bars = plt.barh(df['Category'], df['Value'], color=colors)

plt.title('Unit Tardiness during Annual Operations', fontsize=16, weight='bold')
plt.xlabel('Tardiness')

# Add label ONLY to the maximum bar
bar = bars[max_index]
plt.text(
    max_value + (max_value * 0.03),   # small offset to the right
    bar.get_y() + bar.get_height() / 2,
    str(max_value),
    va='center',
    fontsize=11,
    fontweight='bold'
)

# Adjust x-axis to prevent label clipping
x_padding = max_value * 0.15
plt.xlim(0, max_value + x_padding)

plt.tight_layout()
plt.show()
```

# Tree Map

```python
import matplotlib.pyplot as plt
import squarify

# Sample data
labels = ['Alpha', 'Bravo', 'Charlie', 'Delta', 'Echo', 'Foxtrot', 'Golf']
sizes = [50, 30, 20, 15, 10, 5, 40]

# Normalize colors using a colormap
import matplotlib.cm as cm
import numpy as np

custom_colors = ['#FFC2C7', '#B6E5D8', '#FBE5C8', '#8FDDE7', '#EAFBFF', '#E2C1DE', '#F7E7E1']

plt.figure(figsize=(10,6))
squarify.plot(
    sizes=sizes,
    label=labels,
    color=custom_colors,
    alpha=0.9,
    pad=False,              # remove padding
    edgecolor="white",      # optional thin border
    linewidth=0.5           # make border slim
)
plt.axis('off')
plt.title("Unit Tardiness during the 2nd Year of Operations", fontsize=16, weight='bold', color='#AB202A')
plt.show()
```

# Doughnut Chart

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
labels = ['Apples', 'Bananas', 'Cherries', 'Dates', 'Elderberries']
sizes = [40, 20, 15, 15, 10]
colors = ['#FFC2C7', '#B6E5D8', '#FBE5C8', '#8FDDE7', '#7F8FA3']

plt.figure(figsize=(8,5))

# Draw the doughnut without labels
wedges, _ = plt.pie(
    sizes,
    labels=None,
    startangle=90,
    colors=colors,
    wedgeprops={'width':0.7, 'edgecolor':'white'}
)

# Total for percentage calculation
total = sum(sizes)

# Place percentages in the center of each slice
for i, wedge in enumerate(wedges):
    theta = (wedge.theta2 + wedge.theta1) / 2.0  # middle angle
    x = 0.7 * np.cos(np.deg2rad(theta))          # radius = 0.7
    y = 0.7 * np.sin(np.deg2rad(theta))
    plt.text(x, y, f"{sizes[i]/total*100:.1f}%", ha='center', va='center', color='white', fontsize=12, fontweight='bold')

# Add legend on the right
plt.legend(wedges, labels, title="Fruits", loc="center left", bbox_to_anchor=(1, 0.5), fontsize=12)

plt.title("Types of Fruits Bought (Percentages)", fontsize=16, weight='bold')
plt.axis('equal')  # Keep circular
plt.show()
```

# Line Chart

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug'],
    'Sales': [150, 200, 180, 220, 300, 280, 350, 400]
}

df = pd.DataFrame(data)

# Identify max value
max_index = df['Sales'].idxmax()
max_value = df.loc[max_index, 'Sales']

plt.figure(figsize=(10,6))

# Plot line
plt.plot(
    df['Month'],
    df['Sales'],
    marker='o',
    color='#4e79a7',
    linewidth=3,
    markersize=8,
    linestyle='-',
    alpha=0.9
)

# Fill area below the line
plt.fill_between(
    df['Month'],
    df['Sales'],
    color='#4e79a7',
    alpha=0.2  # transparency
)

# Add data labels
for i, value in enumerate(df['Sales']):
    plt.text(i, value + 5, str(value), ha='center', va='bottom', fontsize=10, fontweight='bold')

# Styling
plt.title('Monthly Sales', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.grid(alpha=0.3)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)

#  Adjust Y-axis so the label never touches the top and it starts at zero
y_padding = max_value * 0.15   # 15% headroom (adjust as you like)
plt.ylim(0, max_value + y_padding)
plt.ylim(bottom=0)

plt.tight_layout()
plt.show()
```

```python
import folium
import pandas as pd

# Corrected DataFrame with proper US city coordinates
df = pd.DataFrame({
    'Area': ['New York', 'Chicago', 'Minneapolis', 'Houston', 'Boston'],
    'Latitude': [40.7128, 41.8781, 44.9778, 29.7604, 42.3601],
    'Longitude': [-74.0060, -87.6298, -93.2650, -95.3698, -71.0589],
    'Value': [300, 250, 400, 150, 200]
})

# Create Folium map (initial location is roughly center of US)
m = folium.Map(location=[39.8283, -98.5795], zoom_start=5, tiles='CartoDB positron')

# Add circle markers with tooltips
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Value'] / 20,   # scale marker size by value
        color='blue',
        fill=True,
        fill_color='cyan',
        fill_opacity=0.6,
        popup=f"{row['Area']}: {row['Value']}",
        tooltip=folium.Tooltip(f"{row['Area']}: {row['Value']}", sticky=True)
    ).add_to(m)

# Automatically adjust map to include all markers
bounds = [[df['Latitude'].min(), df['Longitude'].min()],
          [df['Latitude'].max(), df['Longitude'].max()]]
m.fit_bounds(bounds)

# Display map in Jupyter notebook
m.save("choropleth_map1.html")
[Open Interactive Map](choropleth_map1.html)
```

```python
import folium
import pandas as pd
import branca.colormap as cm

# Sample DataFrame
df = pd.DataFrame({
    'Area': ['New York', 'Chicago', 'Minneapolis', 'Houston', 'Boston'],
    'Latitude': [40.7128, 41.8781, 44.9778, 29.7604, 42.3601],
    'Longitude': [-74.0060, -87.6298, -93.2650, -95.3698, -71.0589],
    'Value': [520, 350, 400, 450, 300]
})

# Create color scale
min_val = df['Value'].min()
max_val = df['Value'].max()
colormap = cm.linear.RdYlBu_09.scale(min_val, max_val)
colormap.caption = 'Value'

# Create Folium map
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4, tiles='CartoDB positron')

# Add circle markers with colors based on value
for _, row in df.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Value'] / 10,  # size scaled by value
        color=colormap(row['Value']),
        fill=True,
        fill_color=colormap(row['Value']),
        fill_opacity=0.3
    ).add_to(m)
    
    # Always-visible label
    folium.map.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=folium.DivIcon(
            html=f"""<div style="font-size:7px; font-weight:bold; color:black">{row['Value']}</div>"""
        )
    ).add_to(m)

# Add the color legend
colormap.add_to(m)

# Fit map bounds
bounds = [[df['Latitude'].min(), df['Longitude'].min()],
          [df['Latitude'].max(), df['Longitude'].max()]]
m.fit_bounds(bounds)

# Display map in Jupyter notebook
m.save("choropleth_map2.html")
[Open Interactive Map](choropleth_map2.html)
```

# Color Choices

### **Sequential palettes (good for values from low â†’ high)**

* `'YlOrRd'` â†’ Yellow â†’ Orange â†’ Red
* `'YlGn'` â†’ Yellow â†’ Green
* `'YlGnBu'` â†’ Yellow â†’ Green â†’ Blue
* `'OrRd'` â†’ Orange â†’ Red
* `'PuRd'` â†’ Purple â†’ Red
* `'BuPu'` â†’ Blue â†’ Purple
* `'GnBu'` â†’ Green â†’ Blue

### **Diverging palettes (good for values around a midpoint)**

* `'RdBu'` â†’ Red â†” Blue
* `'RdYlBu'` â†’ Red â†” Yellow â†” Blue
* `'PuOr'` â†’ Purple â†” Orange
* `'BrBG'` â†’ Brown â†” Blue-Green

```python
import folium
import pandas as pd

# Corrected DataFrame with proper US city coordinates
df = pd.DataFrame({
    'Area': ['New York', 'Chicago', 'Minneapolis', 'Houston', 'Boston'],
    'Latitude': [40.7128, 41.8781, 44.9778, 29.7604, 42.3601],
    'Longitude': [-74.0060, -87.6298, -93.2650, -95.3698, -71.0589],
    'Value': [300, 250, 400, 150, 200]
})

# Create Folium map
m = folium.Map(location=[39.8283, -98.5795], zoom_start=4, tiles='CartoDB positron')

# Add circle markers and always-visible labels
for _, row in df.iterrows():
    # Circle marker
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=row['Value'] / 20,
        color='blue',
        fill=True,
        fill_color='cyan',
        fill_opacity=0.6
    ).add_to(m)
    
    # Always-visible label using DivIcon
    folium.map.Marker(
        location=[row['Latitude'], row['Longitude']],
        icon=folium.DivIcon(
            html=f"""<div style="font-size:7px; font-weight:bold; color:black">{row['Value']}</div>"""
        )
    ).add_to(m)

# Automatically adjust map to include all markers
bounds = [[df['Latitude'].min(), df['Longitude'].min()],
          [df['Latitude'].max(), df['Longitude'].max()]]
m.fit_bounds(bounds)

# Display map in Jupyter notebook
m.save("choropleth_map3.html")
[Open Interactive Map](choropleth_map3.html)
```
